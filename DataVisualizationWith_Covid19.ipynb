## Data Visualization with Corona Dataset

## Import-Load Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('corona_dat.csv')
```

```python
#Daily case numbers in Turkey
df[['date','Turkey']]
```

```python
#Daily case numbers in selected countries
df[['date','Turkey','Italy','Germany','China','US', 'United Kingdom','Japan']]
```

```python
#Put Turkey in a separate series
df_tr=df['Turkey']
type(df_tr)
```

```python
#Draw the daily case change in turkey
df_tr.plot()
```

```python
#Plot the daily case count change in selected countries
df_countries=df[['Turkey','Italy','Germany','China','US','United Kingdom','Japan']]
df_countries.plot()
```

```python
#Total number of cases in selected countries
df_countries.sum(axis=0)
```

```python
#Total number of cases
df_countries.sum(axis=0).plot(kind='bar')
```

```python
sum_daily=df_countries.cumsum(axis=0) # add the numbers in the same column one under the other
sum_daily
```

```python
sum_daily['Countries_Total']=sum_daily.sum(axis=1)
sum_daily
```

```python
sum_daily.plot()
```

```python
df_4=df[['date','Turkey','Italy','Germany','China','US','United Kingdom','Japan']]
df_4.info()
```

```python
df_4
```

```python
#Since it is historically sorted, we can set date as index.
df_4['date']=pd.to_datetime(df_4['date'])
#df_4['month']=pd.DatetimeIndex(df_4['date']).month
#df['month'] = df['Date'].dt.month
df_4.set_index('date',inplace=True) #Set the index as date and index by date
df_4
```

```python
#Make the number of cases a cumulative total
df_4.cumsum(axis=0).plot()
```

## Live Charting

```python
df_yarisan=df_4.cumsum(axis=0) # Create a dataset on cumulative totals of daily cases
df_yarisan
```

```python
sns.lineplot(data=df_yarisan)
```
